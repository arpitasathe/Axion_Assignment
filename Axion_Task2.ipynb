```python
def generate_tags(row):
    tags = []
    if row['TOTALCOST'] > 10000:
        tags.append('HighRepairCost')
    if row['REPAIR_AGE'] > 500:
        tags.append("HighVehicleAge")
    if row['CAUSAL_PART_NM'] == 'ENGINE':
      tags.append('EngineFailure')
    if row['PLATFORM'] == 'A':
        tags.append('PlatformA')
    tags.append(row['BODY_STYLE'])

    return ','.join(tags)

df['Tags'] = df.apply(generate_tags, axis=1)

tag_counts = df['Tags'].str.get_dummies(sep=',').sum().sort_values(ascending=False)
print("Tag Counts:\n", tag_counts)

high_cost_repairs = df[df['Tags'].str.contains('HighRepairCost')]

print("\nBody Style vs. Failure Counts:")
body_style_failure = df.groupby('BODY_STYLE')['Tags'].apply(lambda x: x.str.split(',').explode().value_counts())
body_style_failure
```
